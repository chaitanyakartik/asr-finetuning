```python
from datasets import get_dataset_config_names, load_dataset
import pandas as pd

# The 4 datasets we need to inspect
DATASET_IDS = [
    "ai4bharat/Shrutilipi",
    "ai4bharat/kathbath",
    "ai4bharat/IndicVoices",
    "ARTPARK-IISc/Vaani",
    "ARTPARK-IISc/Vaani-transcription-part"
]

def inspect_datasets():
    print("--- 🕵️‍♀️ Running Deep Inspection ---")

    for ds_id in DATASET_IDS:
        print(f"\n[{ds_id}]")
        try:
            # 1. FETCH CONFIGS (Don't guess, ask the API)
            configs = get_dataset_config_names(ds_id)

            # 2. FILTER FOR KANNADA
            # We look for common codes: 'kn', 'kannada', or 'Karnataka' (for Vaani)
            # This is just a search filter, not a hardcoded config.
            candidates = [c for c in configs if c in ['kn', 'kannada','Kannada'] or 'Karnataka' in c]

            if not candidates:
                print(f"   ⚠️ No obvious Kannada config found in: {configs[:5]}... (Total {len(configs)})")
                continue

            # Pick the first candidate to test (usually the specific config we need)
            target_config = candidates[0]
            print(f"   ✅ Found Config: '{target_config}' (and {len(candidates)-1} others)")

            # 3. STREAM ONE ROW
            # Now we use the EXACT config we just found
            ds = load_dataset(ds_id, target_config, split="train", streaming=True, trust_remote_code=True)
            sample = next(iter(ds))

            # 4. REPORT STRUCTURE
            # We don't assume column names. We print what exists.
            print(f"   👀 Sample Keys: {list(sample.keys())}")

            # Check Audio
            if 'audio' in sample:
                print(f"   🎤 Audio SR: {sample['audio']['sampling_rate']}")
            elif 'audio_filepath' in sample: # Some use this
                 print(f"   🎤 Audio Object: {type(sample['audio_filepath'])}")

            # Check Text (Print the first 50 chars to verify it's Kannada)
            # We check common text keys dynamically
            text_val = "Not Found"
            for key in ['text', 'sentence', 'transcription', 'transcript']:
                if key in sample:
                    text_val = sample[key]
                    break
            print(f"   📝 Text Sample: {str(text_val)[:50]}...")

        except Exception as e:
            print(f"   ❌ ERROR: {e}")

if __name__ == "__main__":
    inspect_datasets()
```

```python
import os
import json
import soundfile as sf
from datasets import load_dataset, Audio
from tqdm import tqdm

# --- CONFIGURATION ---
BASE_DIR = "raw_simulation"
SAMPLES_TO_FETCH = 10

# ✅ CORRECTED CONFIGS
DATASETS = [
    # 1. Shrutilipi
    {
        "name": "Shrutilipi",
        "path": "ai4bharat/Shrutilipi",
        "config": "kannada",
        "split": "train",
        "audio_col": "audio_filepath",
        "text_col": "text"
    },

    # 2. Kathbath
    {
        "name": "Kathbath",
        "path": "ai4bharat/kathbath",
        "config": "kannada",
        "split": "train",
        "audio_col": "audio_filepath",
        "text_col": "text"
    },

    # 3. IndicVoices
    {
        "name": "IndicVoices",
        "path": "ai4bharat/IndicVoices",
        "config": "kannada",
        "split": "train",
        "audio_col": "audio_filepath",
        "text_col": "text"
    },

    # 4. Vaani (Clean Transcriptions)
    # Replaced as requested with the verified transcribed subset
    {
        "name": "Vaani",
        "path": "ARTPARK-IISc/Vaani-transcription-part",
        "config": "Kannada",  # Confirmed via your log
        "split": "train",
        "audio_col": "audio",       # Confirmed via your log
        "text_col": "transcript"    # Confirmed via your log
    },
]

def simulate_raw_download():
    print("--- 🚀 Starting Simulation: Downloading First 10 Samples ---")

    for ds_conf in DATASETS:
        print(f"\nProcessing: {ds_conf['name']}...")

        # Create folder
        ds_dir = os.path.join(BASE_DIR, ds_conf['name'])
        os.makedirs(ds_dir, exist_ok=True)

        try:
            # 1. Load Stream
            ds = load_dataset(
                ds_conf['path'],
                ds_conf['config'],
                split=ds_conf['split'],
                streaming=True
            )

            raw_metadata = []

            # 2. Stream Data
            for i, item in enumerate(ds.take(SAMPLES_TO_FETCH)):

                # --- AUDIO HANDLING ---
                audio_key = ds_conf['audio_col']
                audio_item = item[audio_key]

                try:
                    # Standard HF streaming dict format
                    if isinstance(audio_item, dict) and 'array' in audio_item:
                        audio_array = audio_item['array']
                        orig_sr = audio_item['sampling_rate']
                    else:
                        # Fallback for some object types
                        audio_array = audio_item['array']
                        orig_sr = audio_item['sampling_rate']
                except Exception as e:
                    print(f"   ⚠️ Audio format issue for item {i}: {e}. Trying generic 'audio' key...")
                    # Fallback generic access
                    audio_array = item['audio']['array']
                    orig_sr = item['audio']['sampling_rate']

                # --- TEXT HANDLING ---
                text_content = item.get(ds_conf['text_col'], "")

                # --- SAVE TO DISK ---
                filename = f"{ds_conf['name']}_raw_{i}.wav"
                filepath = os.path.join(ds_dir, filename)
                sf.write(filepath, audio_array, orig_sr)

                # --- LOG METADATA ---
                raw_metadata.append({
                    "filename": filename,
                    "original_sr": orig_sr,
                    "raw_text": text_content,
                    "original_path": os.path.abspath(filepath),
                    "source_config": ds_conf['config']
                })

            # Save Metadata JSON
            meta_path = os.path.join(ds_dir, "raw_metadata.json")
            with open(meta_path, "w", encoding="utf-8") as f:
                json.dump(raw_metadata, f, indent=4, ensure_ascii=False)

            print(f"✅ Saved 10 samples to: {ds_dir}")

        except Exception as e:
            print(f"❌ Failed {ds_conf['name']}: {e}")

if __name__ == "__main__":
    simulate_raw_download()
```

```python
import os
import json
import glob
import pandas as pd

# --- CONFIGURATION ---
BASE_DIR = "raw_simulation"

def run_eda_on_simulation():
    print("--- 🔍 Step 2: Exploratory Analysis (Inspection) ---")

    # 1. Find all the metadata.json files we created
    # We look for */raw_metadata.json inside raw_simulation
    meta_files = glob.glob(os.path.join(BASE_DIR, "*", "raw_metadata.json"))

    if not meta_files:
        print("❌ No metadata files found! Did Step 1 finish successfully?")
        return

    for meta_file in meta_files:
        # Get dataset name from folder path
        dataset_name = os.path.basename(os.path.dirname(meta_file))

        print(f"\n{'='*40}")
        print(f"📊 DATASET: {dataset_name}")
        print(f"{'='*40}")

        try:
            # 2. Load the JSON
            with open(meta_file, 'r', encoding='utf-8') as f:
                data = json.load(f)

            # Convert to Pandas for easy stats
            df = pd.DataFrame(data)

            # 3. Check Sampling Rate (The most critical check)
            unique_srs = df['original_sr'].unique()
            print(f"🔹 Sampling Rates Found: {unique_srs}")

            if len(unique_srs) > 1:
                print(f"   ⚠️ WARNING: Mixed sampling rates! {unique_srs}")
            elif 16000 not in unique_srs:
                print(f"   ⚠️ ACTION: Needs resampling (Current: {unique_srs[0]} Hz)")
            else:
                print(f"   ✅ Audio is clean 16kHz.")

            # 4. Check Text Quality
            # Count how many are empty or null
            empty_text_count = df['raw_text'].isnull().sum() + (df['raw_text'] == "").sum()
            print(f"🔹 Empty/Null Transcripts: {empty_text_count} / {len(df)}")

            # 5. Show a sample
            # We grab the first non-empty text to see if it looks like Kannada
            sample_text = df['raw_text'].dropna().iloc[0] if not df['raw_text'].dropna().empty else "NO TEXT FOUND"
            print(f"🔹 Text Sample: \"{sample_text[:60]}...\"")

        except Exception as e:
            print(f"❌ Error analyzing {dataset_name}: {e}")

if __name__ == "__main__":
    run_eda_on_simulation()
```

```python
import os
import json
import shutil
import soundfile as sf
from tqdm import tqdm

# --- CONFIGURATION ---
DATASET_NAME = "Shrutilipi"
# 1. INPUT: Where the raw wavs and metadata.json are sitting right now
INPUT_DIR = f"raw_simulation/{DATASET_NAME}"

# 2. OUTPUT: Where you want the final clean dataset to live
OUTPUT_DIR = f"processed_data/{DATASET_NAME}"
FINAL_WAV_DIR = os.path.join(OUTPUT_DIR, "wavs")

# Ensure output directories exist
os.makedirs(FINAL_WAV_DIR, exist_ok=True)

def pipeline_shrutilipi():
    print(f"--- 🏭 Starting NeMo Pipeline for {DATASET_NAME} ---")

    # 1. Load the Inventory (metadata.json from the download step)
    meta_path = os.path.join(INPUT_DIR, "raw_metadata.json")
    if not os.path.exists(meta_path):
        print(f"❌ Critical Error: Metadata not found at {meta_path}")
        return

    with open(meta_path, 'r', encoding='utf-8') as f:
        raw_inventory = json.load(f)

    manifest_entries = []

    print(f"   Processing {len(raw_inventory)} files...")

    for item in tqdm(raw_inventory):
        try:
            # --- A. LOCATE FILE ---
            # In simulation, we used absolute paths, but let's be safe and check relative too
            source_path = item.get('original_path')
            if not os.path.exists(source_path):
                # Fallback: look in the input dir directly
                source_path = os.path.join(INPUT_DIR, item['filename'])

            if not os.path.exists(source_path):
                print(f"   ⚠️ Missing File: {item['filename']}")
                continue

            # --- B. VERIFY AUDIO (The 'Gatekeeper') ---
            # We check if it is valid 16kHz audio
            try:
                info = sf.info(source_path)
            except Exception:
                print(f"   ⚠️ Corrupt Audio: {item['filename']}")
                continue

            if info.samplerate != 16000:
                print(f"   ⚠️ Wrong SR ({info.samplerate}Hz): {item['filename']} - Skipping")
                # In a real production run, you would add resample logic here.
                # For now, we skip to keep the pipeline strict.
                continue

            # --- C. FINALIZE ---
            # Copy to the clean 'wavs' folder
            # We rename it to be safe and unique
            new_filename = f"shrutilipi_{item['filename']}" if not item['filename'].startswith("Shrutilipi") else item['filename']
            dest_path = os.path.join(FINAL_WAV_DIR, new_filename)

            # Copy the file (Fast)
            shutil.copy2(source_path, dest_path)

            # --- D. MANIFEST ENTRY ---
            # NeMo requires absolute paths in the JSON
            entry = {
                "audio_filepath": os.path.abspath(dest_path),
                "text": item['raw_text'],
                "duration": info.duration,
                "lang": "kn",
                "source": "shrutilipi_news"
            }
            manifest_entries.append(entry)

        except Exception as e:
            print(f"   ❌ Unexpected Error on {item['filename']}: {e}")

    # --- E. SAVE MANIFEST ---
    # We save in JSONL (JSON Lines) format
    manifest_path = os.path.join(OUTPUT_DIR, "train_manifest.json")
    with open(manifest_path, "w", encoding="utf-8") as f:
        for entry in manifest_entries:
            json.dump(entry, f, ensure_ascii=False)
            f.write('\n')

    print(f"\n✅ Pipeline Success!")
    print(f"   📂 Clean Audio: {FINAL_WAV_DIR}")
    print(f"   📄 NeMo Manifest: {manifest_path}")
    print(f"   📊 Total Processed: {len(manifest_entries)} / {len(raw_inventory)}")

if __name__ == "__main__":
    pipeline_shrutilipi()
```

```python
import os
import json
import shutil
import soundfile as sf
from tqdm import tqdm

# --- CONFIGURATION ---
BASE_INPUT_DIR = "raw_simulation"
BASE_OUTPUT_DIR = "processed_data"

# We map the specific text keys we found during inspection
DATASET_CONFIGS = [
    {
        "name": "Kathbath",
        "text_key": "text",       # Confirmed in EDA
        "sub_source": "kb_clean"
    },
    {
        "name": "IndicVoices",
        "text_key": "text",       # Confirmed in EDA
        "sub_source": "iv_natural"
    },
    {
        "name": "Vaani",
        "text_key": "text",       # Confirmed in EDA (mapped from transcript in step 1)
        "sub_source": "vaani_dialect"
    }
]

def run_unified_pipeline():
    print("--- 🏭 Starting Unified NeMo Pipeline ---")

    for ds_conf in DATASET_CONFIGS:
        ds_name = ds_conf['name']
        print(f"\nProcessing: {ds_name}...")

        # 1. Setup Paths
        input_dir = os.path.join(BASE_INPUT_DIR, ds_name)
        output_dir = os.path.join(BASE_OUTPUT_DIR, ds_name)
        final_wav_dir = os.path.join(output_dir, "wavs")
        os.makedirs(final_wav_dir, exist_ok=True)

        # 2. Load Metadata
        meta_path = os.path.join(input_dir, "raw_metadata.json")
        if not os.path.exists(meta_path):
            print(f"   ⚠️ Skipping {ds_name} (No metadata found at {meta_path})")
            continue

        with open(meta_path, 'r', encoding='utf-8') as f:
            inventory = json.load(f)

        manifest_entries = []

        # 3. Process Files
        for item in tqdm(inventory):
            try:
                # Locate Source
                source_path = item.get('original_path')
                # Fallback check
                if not source_path or not os.path.exists(source_path):
                     source_path = os.path.join(input_dir, item['filename'])

                if not os.path.exists(source_path):
                    continue

                # Verify Audio (16kHz Check)
                info = sf.info(source_path)
                if info.samplerate != 16000:
                    print(f"   ⚠️ Skipping {item['filename']} (SR: {info.samplerate})")
                    continue

                # Copy to Final
                # We prefix filename to avoid collisions if we merge folders later
                new_filename = f"{ds_name.lower()}_{item['filename']}"
                dest_path = os.path.join(final_wav_dir, new_filename)
                shutil.copy2(source_path, dest_path)

                # Create Manifest Entry
                # Note: We use the key from config, but your simulation script
                # might have normalized them all to 'raw_text' or 'text' in the json.
                # Let's check the JSON structure you pasted earlier:
                # It used keys like "raw_text". Let's use that.
                text_content = item.get('raw_text', "")

                if not text_content:
                    continue

                entry = {
                    "audio_filepath": os.path.abspath(dest_path),
                    "text": text_content,
                    "duration": info.duration,
                    "lang": "kn",
                    "source": ds_conf['sub_source']
                }
                manifest_entries.append(entry)

            except Exception as e:
                print(f"   ❌ Error: {e}")

        # 4. Save Manifest
        manifest_path = os.path.join(output_dir, "train_manifest.json")
        with open(manifest_path, "w", encoding="utf-8") as f:
            for entry in manifest_entries:
                json.dump(entry, f, ensure_ascii=False)
                f.write('\n')

        print(f"   ✅ Done! Manifest: {manifest_path}")

if __name__ == "__main__":
    run_unified_pipeline()
```
